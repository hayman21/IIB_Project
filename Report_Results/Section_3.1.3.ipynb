```python
import numpy as np
import matplotlib.pyplot as plt
from Base_Files.ClassLevyJumpProcesses import TemperedStableSubordinator
from Base_Files.ClassPlotter import Plotter


def path_simulation(t1=0.0, t2=1.0, num_obs=500, num_epochs=2000, subordinator_truncation=0.0, kappa=0.5, delta=2.1,  gamma=1.25, nProcesses=10):
    time_ax = np.linspace(t1, t2, num_obs)
    ax = plt.subplot()
    g_sub = TemperedStableSubordinator(t1, t2, num_obs, num_epochs, subordinator_truncation, kappa=kappa, delta=delta, gamma=gamma)
    title = "Sample Tempered Stable Paths with $\\kappa, \\delta, \\gamma = " + str(kappa) + " ," + str(delta) + " ," + str(gamma) + " $ at time $t = 1$"
    for i in range(nProcesses):
        plt.plot(time_ax, g_sub.generate_path())
    plt.xlabel("Time")
    plt.ylabel("Location")
    #plt.title("Realisations of Tempered Stable Paths with $\\kappa, \\delta, \\gamma = " + str(kappa) + " ," + str(delta) + " ," + str(gamma))
    plt.show()
```
